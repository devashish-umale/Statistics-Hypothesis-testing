```python
import pandas as pd 
import numpy as np
farm= pd.read_csv('DS1_C5_S6_FarmSize_Data_Concept (1).csv')
farm
```

```python
farm.info()
```

```python
sample=farm.sample(35, random_state=10)
sample
```

```python
mean_sample=sample['2019_Average_farm_size(acres)'].mean()
std_sample=sample['2019_Average_farm_size(acres)'].std()
pop_mean= farm['2019_Average_farm_size(acres)'].mean()
```

```python
mean_sample, std_sample
```

```python
z_score=(mean_sample-pop_mean)/(std_sample/35**.5)
```

```python
from scipy.stats import norm
p_val=norm.sf(z_score)
p_val
```

```python
if p_val < 0.05:
 print('we reject null hypothsis')
else :
    print('we accept ')
```

```python
city=pd.read_csv('DS1_C5_S6_CityPrice_Data_Challenge (1).csv')
city
```

```python
cal=city['California']
flor=city['Florida']
```

```python
flor.dropna(inplace=True)
```

```python
mean_cal=cal.mean()
mean_flor=flor.mean()

var_cal=cal.var()
var_flor=flor.var()

n_cal=len(cal)
n_flor=len(flor)
```

```python
import math
z_score=((mean_cal-mean_flor)-0)/ math.sqrt(abs((var_cal/n_cal)-(var_flor/n_flor)))
z_score
```

```python
p_val=norm.sf(abs(z_score))*2
p_val
```

```python
if p_val < 0.025:
 print('we reject null hypothsis')
else :
    print('we accept ')
```

```python
# chi square teat for comparing two categorical

df=pd.DataFrame({"subject":['maths','physics','maths','maths','physics','physics'],"grade":['a','b','a','b','b','a']})
df
```

```python
cross=pd.crosstab(df.subject ,df.grade)
cross
```

```python
from scipy.stats import chi2_contingency

p_val=chi2_contingency(cross)[1]
p_val
```

```python
if p_val < 0.05:
 print('we reject null hypothsis')
else :
    print('we accept ')
```

```python
#One Way Annova for three
li1=[1,4,2,5,2,3,5,6,2,5,6]
li2=[11,4,2,15,2,13,5,16,2,5,16]
li3=[21,4,12,22,22,34,25,16,24,5,5]

# ho="there is no significant association between list1,list2 and list3"
# ha="there is significant association between list1,list2 and list3
```

```python
if p_val < 0.025:
 print('we reject null hypothsis')
else :
    print('we accept alternate ')
```

```python
# 2 way annova for one numerical column on two categorical
##Two Way Annova
data=pd.DataFrame({"gender":['M','M','M','F','F','M','F','F','F','M','F'],
                  "stan":['a','a','b','a','a','b','b','b','a','b','a'],
                  "marks":[20,13,13,20,34,21,60,32,36,32,21]})
data

## ho="There is no significant difference on marks,gender and stan values"
## ha="There is a significant difference based on marks,gender and stan value"
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model= ols('marks~C(gender)+C(stan)+C(gender):C(stan)', data=data).fit()
annova_tab= sm.stats.anova_lm(model, type=2)
print(annova_tab)
```

```python
if annova_tab['PR(>F)'][2] < 0.05:
 print('we reject null hypothsis')
else :
    print('we accept alternate hypothisis ')
```
